```python
import matplotlib.pyplot as plt
import numpy as np
def function(x):
    return x**2
learning_rate=0.1
x_current=3
iterations=20
x_values=[x_current]
y_values=[function(x_current)]
for _ in range(iterations):
    gradient=2*x_current
    x_new=x_current-learning_rate*gradient
    x_current=x_new
    x_values.append(x_current)
    y_values.append(function(x_current))
x=np.linspace(-3.5,3.5,500)
y=function(x)
plt.figure(figsize=(6,6))
plt.plot(x,y,label="y=x**2",color="blue")
plt.scatter(x_values,y_values,color='red',label="Gradient Descent Steps")
plt.plot(x_values,y_values,color='red',linestyle='--',label="path of descent")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def function(x):
    return x**4-3*x**3+2
learning_rate=0.01
x_current=3
iterations=20
x_values=[x_current]
y_values=[function(x_current)]
for _ in range(iterations):
    gradient=4*x_current**3-9*x_current**2
    x_new=x_current-learning_rate*gradient
    x_current=x_new
    x_values.append(x_current)
    y_values.append(function(x_current))
x=np.linspace(-1,3,100)
y=function(x)
plt.figure(figsize=(6,6))
plt.plot(x,y,label="y=x**4-3*x**3+2",color="blue")
plt.scatter(x_values,y_values,color='red',label="Gradient Descent Steps")
plt.plot(x_values,y_values,color='red',linestyle='--',label="path of descent")
plt.legend()
plt.show()
```
